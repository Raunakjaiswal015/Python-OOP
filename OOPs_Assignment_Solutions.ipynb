# 1. What is Object-Oriented Programming (OOP)?
       
    A programming paradigm that models software as interacting objects, each bundling state (data/attributes) and behavior (methods). It emphasizes encapsulation, abstraction, inheritance, and polymorphism.

# 2. What is a class in OOP?

    A blueprint for creating objects. It defines the attributes and methods that the instantiated objects will have.

# 3. What is an object in OOP?

    A concrete instance of a class with its own copy of data (attribute values) and access to the class’s methods.

# 4. Difference between abstraction and encapsulation

Abstraction: Hides conceptual complexity—exposes only the essential features (e.g., an interface).

Encapsulation: Hides internal data/implementation by bundling data with methods and restricting direct access (e.g., name-mangling/private attributes).

# 5. What are dunder methods in Python?

    “Double-underscore” (magic) methods like _init, __str, __repr, __len, __add_ that let objects integrate with Python syntax/protocols (construction, printing, operators, iteration, etc.).

# 6. Explain inheritance

    A class (child/subclass) can derive from another class (parent/base) and reuse/extend/override its behavior, enabling hierarchical relationships and code reuse.

# 7. What is polymorphism in OOP?

    The same interface (method name) can work with different underlying types/implementations, e.g., multiple classes implementing draw() differently.

# 8. How is encapsulation achieved in Python?

    Convention: prefix _attr for “internal” use.

    Name-mangling: _attr becomes _ClassName_attr to discourage external access.

    Use properties/getters/setters to control access.

# 9. What is a constructor in Python?

    _init(self, ...) initializes a newly created object’s state immediately after memory allocation (new_). Called automatically on instantiation.

# 10. What are class and static methods?

    @classmethod: Receives cls; used for alternative constructors or behavior tied to the class as a whole.

    @staticmethod: No self/cls; utility function kept inside the class namespace for cohesion.

# 11. What is method overloading in Python?

    Python doesn’t support traditional compile-time overloading by signature. Common patterns: default parameters, args/*kwargs, or functools.singledispatch to emulate behavior.

# 12. What is method overriding in OOP?

    A subclass provides its own implementation of a method with the same name/signature as in the superclass, changing or extending behavior.

# 13. What is a property decorator in Python?

    @property turns a method into a managed attribute: enables computed attributes with optional @attr.setter/@attr.deleter, while keeping attribute-style access.

# 14. Why is polymorphism important?

    It decouples code from concrete types, enabling extensibility, cleaner APIs, and easier testing—clients depend on behavior (interface), not specific classes.
# 15. What is an abstract class in Python?

    A class that can’t be instantiated directly and may declare abstract methods using abc.ABC and @abstractmethod. Subclasses must implement the abstract methods.

# 16. Advantages of OOP

    Modularity, reuse via inheritance/composition, easier maintenance, clear modeling of real-world entities, extensibility, and testability.

# 17. Difference between a class variable and an instance variable

    Class variable: Shared across all instances (Class.var).

    Instance variable: Unique per object (self.var), usually set in _init_.

# 18. What is multiple inheritance in Python?

    A class inherits from more than one base class: class C(A, B): .... Python resolves attribute lookup via the Method Resolution Order (MRO / C3 linearization).

# 19. Purpose of _str_ and _repr_

    _str_: User-friendly, readable string (for print()).

    _repr: Unambiguous, often valid Python expression or detailed developer representation (used in REPL; fallback if __str_ missing).

# 20. Significance of super() in Python

    Accesses the next class in the MRO to call parent implementations without hard-coding the parent name—crucial for cooperative multiple inheritance.

# 21. Significance of the _del_ method

    Finalizer called when an object is about to be garbage-collected; used sparingly (resource cleanup is better handled via context managers/with/_enter/exit_ or try/finally).

# 22. Difference between @staticmethod and @classmethod

    @staticmethod: No automatic first argument; pure utility.

    @classmethod: Receives cls; can manipulate class state and create alternative constructors.

# 23. How does polymorphism work in Python with inheritance?
   
    Dynamic (duck-typed) dispatch: calling obj.method() invokes the most specific override according to the object’s class; different subclasses provide different implementations behind the same method name.

# 24. What is method chaining in Python OOP?

    Design pattern where methods return self (or a new object) to allow consecutive calls: builder.set_a(...).set_b(...).build().

# 25. Purpose of the _call_ method in Python
    
    Makes an instance callable like a function: defining _call_(self, ...) lets you do obj(...), handy for configurable functions, stateful processors, and decorators.

# Q1. Create a class to represent a bank account. Include methods to deposit,withdraw, and check balance.

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return f"Deposited {amount}. New balance: {self.balance}"

    def withdraw(self, amount):
        if amount > self.balance:
            return "Insufficient balance"
        self.balance -= amount
        return f"Withdrew {amount}. New balance: {self.balance}"

    def check_balance(self):
        return f"Current balance: {self.balance}"
```

# Q2. Define a class to represent a student. Include methods to add marks in 3 subjects and calculate the average.

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.marks = []

    def add_marks(self, m1, m2, m3):
        self.marks = [m1, m2, m3]

    def average(self):
        return sum(self.marks) / len(self.marks)
```

# Q3. Create a class Rectangle with methods to calculate area and perimeter.

```python
class Rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area(self):
        return self.length * self.breadth

    def perimeter(self):
        return 2 * (self.length + self.breadth)
```

# Q7. Create a class Circle with radius attribute. Methods to calculate area and circumference.

```python
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def circumference(self):
        return 2 * math.pi * self.radius
```

# Q14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.


```python
# Q14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.
class Employee:
    def __init__(self, hours_worked, hourly_rate):
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, hours_worked, hourly_rate, bonus):
        super().__init__(hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        return super().calculate_salary() + self.bonus
```

# Q16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.




```python
# Q16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        return "Moo"

class Sheep(Animal):
    def sound(self):
        return "Baa"
```
